```python
import requests
from langchain_community.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

loader = TextLoader('/content/sample_text.txt')
document = loader.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
text_chunks = text_splitter.split_documents(documents=document)
```

```python
from langchain.vectorstores import FAISS

vector_store = FAISS.from_documents(text_chunks, embeddings)
```

```python
vector_store
```

```python
# Explicitly create embeddings for text chunks
text_embeddings = embeddings.embed_documents([chunk.page_content for chunk in text_chunks])

# Create FAISS vector store from embeddings and text chunks
vector_store = FAISS.from_embeddings(zip([chunk.page_content for chunk in text_chunks], text_embeddings), embeddings.embed_query)
```

```python
retriever=vector_store.as_retriever()
```

```python
from langchain.prompts import ChatPromptTemplate
```

```python
template="""You are an assistant for question answering tasks.
Use the following pieces of context to answer the question at the end.
If you don't know the answer, just say that you don't know, dont try to make up an answer.
Use % sentences maximum and keep the answers concise.

Question :{question}
Context :{context}
Answer :
"""
```

```python
prompt=ChatPromptTemplate.from_template(template=template)
```

```python
from langchain_google_genai.chat_models import ChatGoogleGenerativeAI
from langchain_core.runnables import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
```

```python
output_parser=StrOutputParser
```

```python
from langchain_google_genai import GoogleGenerativeAI
```

```python
llm_model=GoogleGenerativeAI(api_key=userdata.get('GOOGLE_API_KEY_1'),model="gemini-flash-lite-latest")
```

##Langchain Expression Language

```python
rag_chain=(
    {'context':retriever,'question':RunnablePassthrough()}
    |prompt
    |llm_model
    |output_parser() # Instantiate the output parser
)
```

```python
rag_chain.invoke("How AI is harmful?")
```

```python
from google import genai
from google.genai import types

client = genai.Client()

result = client.models.embed_content(
    model="gemini-embedding-001",
    contents="What is the meaning of life?",
    config=types.EmbedContentConfig(output_dimensionality=768)
)

[embedding_obj] = result.embeddings
embedding_length = len(embedding_obj.values)

print(f"Length of embedding: {embedding_length}")
```

```python
from langchain.embeddings import GoogleGenerativeAIEmbeddings
from langchain.vectorstores import FAISS

# Example documents to index
documents = [
    "LangChain makes it easy to build LLM-powered applications.",
    "FAISS provides fast similarity search for embeddings.",
    "Google Generative AI can create high-quality embeddings for text."
]


# Initialize Google Generative AI Embeddings
# Make sure to have your API key set in environment variables or configure it in the library
embeddings = GoogleGenerativeAIEmbeddings()

# Create embeddings for the documents
doc_embeddings = [embeddings.embed_query(doc) for doc in documents]

# Initialize FAISS and add document embeddings
vector_store = FAISS(embedding_function=embeddings.embed_query, index=None)
vector_store.add_texts(documents)

# Perform a similarity search
query = "How to search similar text using embeddings?"
similar_docs = vector_store.similarity_search(query, k=2)

# Print the results
print("Top similar documents:")
for i, doc in enumerate(similar_docs, start=1):
    print(f"{i}. {doc.page_content}")
```
